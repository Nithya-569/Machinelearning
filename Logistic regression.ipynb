```python
X=df.drop("test_result",axis=1)
```

```python
y=df['test_result']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
scaler=StandardScaler()
X_train_values=scaler.fit_transform(X_train)
X_test_values=scaler.transform(X_test)
```

```python
log_model=LogisticRegression()
```

```python
log_model.fit(X_train_values,y_train)
```

```python
log_model.coef_
```

```python
y_predictions=log_model.predict(X_test_values)
```

```python
y_predictions
```

```python
y_prob_predictions=log_model.predict_proba(X_test_values)
```
